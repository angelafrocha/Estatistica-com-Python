```python
import pandas as pd
from scipy.stats import normaltest
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
dados = pd.read_csv('/content/dados.csv')
dados.shape
```

```python
dados.head(3)
```

```python
# Verificando a normalidade das variáveis quantitativas

fig, ax = plt.subplots(2, 2, figsize=(14, 10))  # 2 linhas, 2 colunas

# Primeiro gráfico: Distribuição das Idades
sns.histplot(dados['Idade'], ax=ax[0, 0], kde=True, edgecolor=None, bins=50)
ax[0, 0].set_title('Distribuição das Idades')
ax[0, 0].set_xlabel('Idade')
ax[0, 0].set_ylabel('Contagem')
ax[0, 0].grid(True)

# Segundo gráfico: Distribuição dos Anos de Estudo
sns.histplot(dados['Anos de Estudo'], ax=ax[0, 1], kde=True, edgecolor=None, bins=50)
ax[0, 1].set_title('Distribuição dos Anos de Estudo')
ax[0, 1].set_xlabel('Anos de estudo')
ax[0, 1].set_ylabel('Contagem')
ax[0, 1].grid(True)

# Terceiro gráfico: Distribuição da Renda
sns.histplot(dados['Renda'], ax=ax[1, 0], kde=True, edgecolor=None, bins=50)
ax[1, 0].set_title('Distribuição da Renda')
ax[1, 0].set_xlabel('Renda')
ax[1, 0].set_ylabel('Contagem')
ax[1, 0].grid(True)

# Quarto gráfico: Distribuição das Alturas
sns.histplot(dados['Altura'], ax=ax[1, 1], kde=True, edgecolor=None, bins=50)
ax[1, 1].set_title('Distribuição das Alturas')
ax[1, 1].set_xlabel('Alturas')
ax[1, 1].set_ylabel('Contagem')
ax[1, 1].grid(True)

plt.tight_layout()  # Ajusta o layout para evitar sobreposição
plt.show()
```

Apenas com os histogramas, nesse caso, é possível verificar que apenas a altura mantém uma distibuição normal

```python
# verificando cada coluna pelo Teste de D'Agostino e Pearson para a Idade
significancia = 0.05
stat_test_idade, p_valor_idade = normaltest(dados.Idade)
print(f'Estatística: {stat_test_idade:.3f}')
print(f'p-valor: {p_valor_idade}')
```

```python
#Rejeição da hipótese H0
p_valor_idade <= significancia
```

O p-valor rejeita a significância, o que implica que distribuição não é normal

```python
# para Anos de estudo
significancia = 0.05

stat_test_aestudos, p_valor_aestudos = normaltest(dados.Renda)

print(f'Estatística: {stat_test_aestudos:.3f}')
print(f'p-valor: {p_valor_aestudos}')

p_valor_aestudos <= significancia
```

```python
# para Anos de Renda
significancia = 0.05

stat_test_renda, p_valor_renda = normaltest(dados.Renda)

print(f'Estatística: {stat_test_renda:.3f}')
print(f'p-valor: {p_valor_renda}')

p_valor_renda <= significancia
```

```python
# para Altura
significancia = 0.05

stat_test_altura, p_valor_altura = normaltest(dados.Altura)

print(f'Estatística: {stat_test_altura:.3f}')
print(f'p-valor: {p_valor_altura}')

p_valor_altura <= significancia
```

Aqui o p-valor não rejeita a significâcia e confirmamos que se trata de uma distribuição normal
